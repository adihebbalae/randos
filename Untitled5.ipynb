```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(0)

# Create datasets
right_skewed = np.random.exponential(scale=2, size=1000)
left_skewed = -np.random.exponential(scale=2, size=1000)
symmetric = np.random.normal(loc=0, scale=1, size=1000)

# Function to plot ogive
def plot_ogive(data, title):
    sorted_data = np.sort(data)
    cumulative = np.arange(1, len(sorted_data) + 1) / len(sorted_data) * 100
    plt.plot(sorted_data, cumulative, marker='o', linestyle='-')
    plt.title(title)
    plt.xlabel('Data Values')
    plt.ylabel('Cumulative Percentage')
    plt.grid(True)
    plt.show()

# Plot ogives
plot_ogive(right_skewed, 'Ogive of Right-Skewed Data')
plot_ogive(left_skewed, 'Ogive of Left-Skewed Data')
plot_ogive(symmetric, 'Ogive of Symmetric Data')
```
